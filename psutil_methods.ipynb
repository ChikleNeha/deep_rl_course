```python
import psutil as ps

ps.cpu_count()
ps.cpu_percent(1)

```

```python
ps.virtual_memory()
```

```python
ps.cpu_count(logical=False)
```

```python
mem = ps.virtual_memory()
print("Total memry : ", mem.total>>30, "GB")
print("Percentage used : ", mem.percent)
```

```python
# Disk and network
disk = ps.disk_usage('/')
print(f"Free Space : {disk.free / (1024**3) }GB")
```

```python
net = ps.net_io_counters()
print(f"Bytes sent : {net.bytes_sent}, Bytes recieved : {net.bytes_recv}")
```

```python
# Processes

all_pids = ps.pids()
print(all_pids)
```

```python
p = ps.Process(72)
print(p.name)
print(p.status)
print(p.memory_info)
```
